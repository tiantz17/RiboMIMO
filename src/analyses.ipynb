## 2. Performance evaluation

```python
list_length = []
list_density = []
list_pred = []
list_fold = []
np.random.seed(RiboMIMONet.seed)
seeds = np.random.randint(0, 1024, RiboMIMONet.num_repeat)
logging.info("{} repeat {} fold cross-validation".format(RiboMIMONet.num_repeat, RiboMIMONet.num_fold))
for repeat in range(RiboMIMONet.num_repeat):
    RiboMIMONet.split_data(seeds[repeat])
    for fold in range(RiboMIMONet.num_fold):
        logging.info("{}/{} repeat {}/{} fold".format(repeat+1, RiboMIMONet.num_repeat, fold+1, RiboMIMONet.num_fold))
        RiboMIMONet.load_model()
        RiboMIMONet.model.load_state_dict(torch.load(model_path+"best_model_{}_{}.pth".format(repeat, fold)))
        RiboMIMONet.index_test = RiboMIMONet.data.list_fold[fold]
        index_eval = RiboMIMONet.index_test
        RiboMIMONet.index_train = RiboMIMONet.data.list_fold[:fold] + RiboMIMONet.data.list_fold[fold+1:]
        RiboMIMONet.index_train = np.concatenate(RiboMIMONet.index_train)
        RiboMIMONet.get_thresh()
        RiboMIMONet.data.get_data_pack(index_eval, RiboMIMONet.add_nt, RiboMIMONet.add_aa)
        RiboMIMONet.data.get_data_pack_cls(index_eval, RiboMIMONet.threshs)
        num_batch = int(np.ceil(len(index_eval) / float(RiboMIMONet.batch_size)))
        temp_density = [[] for _ in range(len(index_eval))]
        temp_pred = [[] for _ in range(len(index_eval))]
        with torch.no_grad():
            for batch in range(num_batch):
                index = np.array(range(batch*RiboMIMONet.batch_size, min((batch+1)*RiboMIMONet.batch_size,len(index_eval))))
                x_input, length, density, mask_density, level, batch_index = RiboMIMONet.get_data_batch(index)
                pred = RiboMIMONet.model.predict(x_input, length)
                length = length.cpu().data.long()
                for i in range(pred.shape[0]):
                    temp_density[batch_index[i]] = density[i][0:length[i]].cpu().data
                    temp_pred[batch_index[i]] = pred[i][0:length[i]].cpu().data
        list_density.extend(temp_density)
        list_pred.extend(temp_pred)
        list_fold.extend([repeat*fold+fold] * len(temp_density))
torch.cuda.empty_cache()
```

```python
list_results = []
for idx, gene in enumerate(RiboMIMONet.data.list_gene):
    list_results.append(RiboMIMONet.model.score_reg(list_pred[idx], list_density[idx]))
list_results = np.array(list_results)
printstring = "\033[91mr2: {:.6f}±{:.6f}, mse: {:.6f}±{:.6f}, corr: {:.6f}±{:.6f}\033[0m"
logging.info("Per-gene correlation, SD across {} genes".format(len(list_results)))
logging.info(printstring.format(np.nanmean(list_results[:,0]), 
                                np.nanstd(list_results[:,0]),
                                np.nanmean(list_results[:,1]), 
                                np.nanstd(list_results[:,1]),
                                np.nanmean(list_results[:,2]), 
                                np.nanstd(list_results[:,2]),))
```

```python
num = 0
for seq in RiboMIMONet.data.dict_seq_codon.values():
    if len(seq) > num:
        num = int(len(seq))

# 10 fold all saliency map
list_distance_cis = np.zeros(num*2)
counts = np.zeros(num*2)
for idx in range(len(RiboMIMONet.data.list_gene)):
    print("{:.2f}%\r".format((1.0+idx)/len(RiboMIMONet.data.list_gene)*100), end="")
    gene_name = RiboMIMONet.data.list_gene[idx]
    seq = RiboMIMONet.data.dict_seq[gene_name]
    len_codon = len(seq)//3
    codons = [seq[i*3:(i+1)*3] for i in range(len_codon)]
    try:
        saliency_sum = pickle.load(open("{}/{}".format(cis_path, gene_name), "rb"))
    except:
        continue
    sm_2d = np.abs(saliency_sum)
    for i in range(5, len_codon-4):
        list_distance_cis[num+i-len_codon+6:num+i-5+1] += sm_2d[5:len_codon-5, i][::-1]
        counts[num+i-len_codon+6:num+i-5+1] += 1
list_distance_cis /= counts

# 10 fold all saliency map
list_distance_cis_sq1 = np.zeros(num*2)
list_distance_cis_sq2 = np.zeros(num*2)
counts1 = np.zeros(num*2)
counts2 = np.zeros(num*2)
for idx in range(len(RiboMIMONet.data.list_gene)):
    print("{:.2f}%\r".format((1.0+idx)/len(RiboMIMONet.data.list_gene)*100), end="")
    gene_name = RiboMIMONet.data.list_gene[idx]
    seq = RiboMIMONet.data.dict_seq[gene_name]
    len_codon = len(seq)//3
    codons = [seq[i*3:(i+1)*3] for i in range(len_codon)]
    try:
        saliency_sum = pickle.load(open("{}/{}".format(cis_path, gene_name), "rb"))
    except:
        continue
    sm_2d = np.abs(saliency_sum)
    for i in range(5, len_codon-4):
        control = list_distance_cis[num+i-len_codon+6:num+i-5+1]
        error = sm_2d[5:len_codon-5, i][::-1]-control
        list_distance_cis_sq1[num+i-len_codon+6:num+i-5+1] += (error**2) * (error > 0)
        list_distance_cis_sq2[num+i-len_codon+6:num+i-5+1] += (error**2) * (error < 0)
        counts1[num+i-len_codon+6:num+i-5+1] += error > 0
        counts2[num+i-len_codon+6:num+i-5+1] += error < 0
counts1[counts1==0] = 1
counts2[counts2==0] = 1
list_distance_cis_sq1 /= counts1
list_distance_cis_sq2 /= counts2
list_distance_cis_sq1[np.isnan(list_distance_cis_sq1)] = 0
list_distance_cis_sq2[np.isnan(list_distance_cis_sq2)] = 0
list_distance_cis_sq1 = np.sqrt(list_distance_cis_sq1)
list_distance_cis_sq2 = np.sqrt(list_distance_cis_sq2)
```

```python
# CAI
import pandas as pd
codon_count = {}
for idx in range(len(RiboMIMONet.data.list_gene)):
    gene_name = RiboMIMONet.data.list_gene[idx]
    seq = RiboMIMONet.data.dict_seq[gene_name]
    len_codon = len(seq)//3
    codon = [seq[i*3:(i+1)*3] for i in range(len_codon)]
    for jdx in range(len_codon):
        if codon[jdx] not in codon_count:
            codon_count[codon[jdx]] = 1
        else:
            codon_count[codon[jdx]] += 1
codon_usage = pd.DataFrame(columns=["Codon", "Number", "AA"])
codon_usage["Codon"] = list(codon_count.keys())
codon_usage["AA"] = [RiboMIMONet.data.codon2aa[codon] for codon in list(codon_count.keys())]
for i in codon_usage.index:
    codon_usage.loc[i, "Number"] = codon_count[codon_usage.loc[i, "Codon"]]
codon_usage_dict = {}
for i in codon_usage.index:
    aa = codon_usage.loc[i, "AA"]
    fraction_max = codon_usage["Number"][codon_usage["AA"]== aa].max()
    codon_usage_dict[codon_usage.loc[i, "Codon"]] = codon_usage.loc[i, "Number"]/fraction_max
list_CAI = np.array([codon_usage_dict[codon] for codon in list_codon])
used_info = "CAI"
```

```python
# CIS
site = "A"
if site == "A":
    site_offset = 0
elif site == "E":
    site_offset = 2
dict_codon_cis = {}
for idx in range(len(RiboMIMONet.data.list_gene)):
    print("{:.2f}%\r".format((1.0+idx)/len(RiboMIMONet.data.list_gene)*100), end="")
    gene_name = RiboMIMONet.data.list_gene[idx]
    seq = RiboMIMONet.data.dict_seq[gene_name]
    len_codon = len(seq)//3
    codons = [seq[i*3:(i+1)*3] for i in range(len_codon)]
    try:
        saliency_sum = pickle.load(open("{}/{}".format(cis_path, gene_name), "rb"))
    except:
        continue
    sm_2d = np.abs(saliency_sum)
    for i in range(saliency_sum.shape[1]-site_offset):
        if codons[i] not in dict_codon_cis:
            dict_codon_cis[codons[i]] = [0, 0]
        dict_codon_cis[codons[i]][0] += sm_2d[:,i+site_offset].sum()
        dict_codon_cis[codons[i]][1] += sm_2d.shape[0]
for codon in dict_codon_cis:
    dict_codon_cis[codon] = dict_codon_cis[codon][0] / dict_codon_cis[codon][1]
list_CIS = np.array([dict_codon_cis[codon] for codon in list_codon])
```

```python
# density
dict_density = {}
for idx in range(len(RiboMIMONet.data.list_gene)):
    gene_name = RiboMIMONet.data.list_gene[idx]
    seq = RiboMIMONet.data.dict_seq[gene_name]
    len_codon = len(seq)//3
    codon = [seq[i*3:(i+1)*3] for i in range(len_codon)]
    density = list_pred[idx].numpy()
    assert len(density) == len(codon)
    for jdx in range(len_codon-site_offset):
        if codon[jdx+site_offset] not in dict_density:
            dict_density[codon[jdx+site_offset]] = [np.abs(density[jdx]), np.array(density[jdx]>0, dtype=int)]
        else:
            dict_density[codon[jdx+site_offset]][0] += np.abs(density[jdx])
            dict_density[codon[jdx+site_offset]][1] += np.array(density[jdx]>0, dtype=int)
for codon in dict_density:
    dict_density[codon] = dict_density[codon][0]/dict_density[codon][1]
list_den = np.array([dict_density[codon] for codon in list_codon])
```

```python
# 10 fold RPF saliency map
window = 10
dict_influence_codon = np.zeros((61, window))
counts = np.zeros((61, window))
for idx in range(len(RiboMIMONet.data.list_gene)):
    print("{:.2f}%\r".format((1.0+idx)/len(RiboMIMONet.data.list_gene)*100), end="")
    gene_name = RiboMIMONet.data.list_gene[idx]
    seq = RiboMIMONet.data.dict_seq[gene_name]
    len_codon = len(seq)//3
    codons = [seq[i*3:(i+1)*3] for i in range(len_codon)]
    try:
        saliency_sum = pickle.load(open("{}/{}".format(cis_path, gene_name), "rb"))
    except:
        continue
    sm_2d = saliency_sum
    for i in range(len_codon):
        temp_codon = codons[i]
        if temp_codon in ["TAA", "TAG", "TGA"]:
            continue
        start = max(0, i-window//2+1)
        stop = min(len_codon, i+window//2+1)
        dict_influence_codon[codon2idx[temp_codon], window//2+1+i-stop:window//2+1+i-start] += sm_2d[start:stop, i][::-1]
        counts[codon2idx[temp_codon], window//2+1+i-stop:window//2+1+i-start] += 1
dict_influence_codon /= counts
```

### 3.4 Long-range contributions

```python
# 10 fold pair saliency map
dict_influence_pair = np.zeros((61, 61))
counts = np.zeros((61, 61))
window = 20
for idx in range(len(RiboMIMONet.data.list_gene)):
    print("{:.2f}%\r".format((1.0+idx)/len(RiboMIMONet.data.list_gene)*100), end="")
    gene_name = RiboMIMONet.data.list_gene[idx]
    seq = RiboMIMONet.data.dict_seq[gene_name]
    len_codon = len(seq)//3
    codons = [seq[i*3:(i+1)*3] for i in range(len_codon)]
    try:
        saliency_sum = pickle.load(open("{}/{}".format(cis_path, gene_name), "rb"))
    except:
        continue
    sm_2d = saliency_sum
    for i, j in np.array(np.triu_indices(len_codon-1, window)).T:
        a = codons[i]
        if a in ["TAA", "TAG", "TGA"]:
            continue
        b = codons[j]
        if b in ["TAA", "TAG", "TGA"]:
            continue
        idxa = codon2idx[a]
        idxb = codon2idx[b]
        # left
        dict_influence_pair[idxa, idxb] += sm_2d[j, i]
        counts[idxa, idxb] += 1
        # right
        dict_influence_pair[idxb, idxa] += sm_2d[i, j]
        counts[idxb, idxa] += 1
dict_influence_pair /= counts
```
